```python
company=soup.find_all('div','companyCardWrapper')
print(company[0].find('div',class_='rating_star_container').text.strip())
```

```python
name=[]
reviews=[]
salaries=[]
interviews=[]
jobs=[]
hq=[]
rating=[]

for i in company:

    name.append(i.find('h2').text.strip())
    reviews.append(i.find('span', class_='companyCardWrapper__ActionCount').text.strip())
    salaries.append(i.find_all('span',class_='companyCardWrapper__ActionCount')[1].text.strip())
    interviews.append(i.find_all('span',class_='companyCardWrapper__ActionCount')[2].text.strip())
    jobs.append(i.find_all('span',class_='companyCardWrapper__ActionCount')[3].text.strip())
    hq.append(i.find('span',class_='companyCardWrapper__interLinking').text.strip())
    rating.append(i.find('div',class_='rating_star_container').text.strip())

print(name)
print(reviews)
print(salaries)
print(interviews)
rating
```

```python
df=pd.DataFrame({'Name':name,
                 'Reviews':reviews,
                 'Salaries':salaries,
                 'Interviews':interviews,
                 'Jobs':jobs,
                 'Head_Quarters':hq,
                 'Rating':rating,
                })
```

-------------------- DataFrame for all the pages ---------------------

```python
final=pd.DataFrame()
for j in range(1,501):
    url=f'https://www.ambitionbox.com/list-of-companies?page={j}'
    webpage = requests.get(url,headers=headers).text
    soup=BeautifulSoup(webpage,'lxml')
    company=soup.find_all('div','companyCardWrapper')
    name=[]
    reviews=[]
    salaries=[]
    interviews=[]
    jobs=[]
    hq=[]
    rating=[]

    for i in company:
        try:
            name.append(i.find('h2').text.strip())
        except:
            name.append(np.nan)
            
        try:
            reviews.append(i.find('span', class_='companyCardWrapper__ActionCount').text.strip())
        except:
            reviews.append(np.nan)

        try:
            salaries.append(i.find_all('span',class_='companyCardWrapper__ActionCount')[1].text.strip())
        except:
            salaries.append(np.nan)

        try:
            interviews.append(i.find_all('span',class_='companyCardWrapper__ActionCount')[2].text.strip())
        except:
            interviews.append(np.nan)


        try:
            jobs.append(i.find_all('span',class_='companyCardWrapper__ActionCount')[3].text.strip())
        except:
            jobs.append(np.nan)


        try:
            hq.append(i.find('span',class_='companyCardWrapper__interLinking').text.strip())
        except:
            hq.append(np.nan)

        try:
            rating.append(i.find('span',class_='companyCardWrapper__interLinking').text.strip())
        except:
            rating.append(np.nan)

    df=pd.DataFrame({'Name':name,
                 'Reviews':reviews,
                 'Salaries':salaries,
                 'Interviews':interviews,
                 'Jobs':jobs,
                 'Head_Quarters':hq,
                 'Rating':rating,
                })
    final = pd.concat([final, df], ignore_index=True)
```
